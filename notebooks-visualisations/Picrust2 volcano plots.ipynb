```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pathabund = pd.read_csv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_B/q2-picrust2_output/pathabun_core_metrics_out/ancom_pathabund_Disease_status_volcano.tsv", sep='\t')
ecmetagen = pd.read_csv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_B/q2-picrust2_output/ecmetagen_core_metrics_out/ancom_ecmetagen_Disease_status_volcano.tsv", sep='\t')
kometagen = pd.read_csv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_B/q2-picrust2_output/kometagen_core_metrics_out/ancom_kometagen_Disease_status_volcano.tsv", sep='\t')
```

```python
value=pathabund['clr']<0
pathabund['color']= np.where(value==True, "up in HC", "up in IBD")
value=ecmetagen['clr']<0
ecmetagen['color']= np.where(value==True, "up in IBD", "up in HC")
value=kometagen['clr']<0
kometagen['color']= np.where(value==True, "up in HC", "up in IBD")
```

```python
color_dict = dict({'up in IBD':'orange',
                   'up in HC':'blue'
                  })
```

```python
clr1 = np.arange(start=-1, stop=1.5, step=0.1)
W1 = np.repeat(320, 25)
clr2 = np.arange(start=-1.5, stop=1.2, step=0.1)
W2 = np.repeat(1786, 27)
clr3 = np.arange(start=-1.5, stop=1.5, step=0.1)
W3 = np.repeat(5622, 30)
```

```python
f1, (ax1,ax2,ax3) = plt.subplots(1, 3, figsize=(30, 10))

sns.scatterplot(ax=ax1, data=pathabund, x='clr', y='W', alpha = 0.6, hue='color', palette=color_dict)
sns.lineplot(ax=ax1, x=clr1, y=W1, linestyle='--', color='black')
ax1.set_xlabel('clr', fontsize=20)
ax1.set_ylabel('W', fontsize=20)
ax1.set_title("MetaCyc pathway abundance", fontsize=20)
ax1.legend(loc=4, prop={'size': 18})

sns.scatterplot(ax=ax2, data=ecmetagen, x='clr', y='W', alpha = 0.6, hue='color', palette=color_dict)
sns.lineplot(ax=ax2, x=clr2, y=W2, linestyle='--', color='black')
ax2.set_xlabel('clr', fontsize=20)
ax2.set_ylabel('W', fontsize=20)
ax2.set_title("Emzyme comission", fontsize=20)
ax2.legend(loc=4, prop={'size': 18})

sns.scatterplot(ax=ax3, data=kometagen, x='clr', y='W', alpha = 0.6, hue='color', palette=color_dict)
sns.lineplot(ax=ax3, x=clr3, y=W3, linestyle='--', color='black')
ax3.set_xlabel('clr', fontsize=20)
ax3.set_ylabel('W', fontsize=20)
ax3.set_title("KEGG Orthology", fontsize=20)
ax3.legend(loc=4, prop={'size': 18})

f1.savefig('/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_B/q2-picrust2_output/volcano-Disease_status.png')

```
