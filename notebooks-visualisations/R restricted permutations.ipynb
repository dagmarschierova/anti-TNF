```r
distance <- read_tsv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_manifest_merged/diversity_HCP_week038/beta/bray_curtis_distance_matrix.tsv")
distance_cd <- read_tsv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_manifest_merged/diversity_HCP_week038-CD/beta/bray_curtis_distance_matrix.tsv")
distance_uc <- read_tsv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_manifest_merged/diversity_HCP_week038-UC/beta/bray_curtis_distance_matrix.tsv")
```

```r
meta_dist <- inner_join(metadata, distance, by="SampleID")
meta_dist_cd <- inner_join(metadata, distance_cd, by="SampleID")
meta_dist_uc <- inner_join(metadata, distance_uc, by="SampleID")
```

```r
meta_dist_cd
```

```r
all_dist <- meta_dist %>%
    select(all_of(.[["SampleID"]])) %>%
    as.dist()

all_dist_cd <- meta_dist_cd %>%
    select(all_of(.[["SampleID"]])) %>%
    as.dist()

all_dist_uc <- meta_dist_uc %>%
    select(all_of(.[["SampleID"]])) %>%
    as.dist()
```

```r
adonis(all_dist~Dg/Week, data=meta_dist)
```

```r
adonis(all_dist~Week, data=meta_dist, permutations=9999)
```

```r
adonis(all_dist~Dg, data=meta_dist, permutations=9999)
```

```r
adonis(all_dist~Dg, data=meta_dist, permutations=how(blocks=meta_dist$Week, nperm=9999))
```

```r
mod <- betadisper(all_dist, meta_dist$Dg)
(mod.HSD <- TukeyHSD(mod))
```

```r
adonis(all_dist_cd~Dg, data=meta_dist_cd, permutations=how(blocks=meta_dist_cd$Week, nperm=9999))
```

```r
mod_cd <- betadisper(all_dist_cd, meta_dist_cd$Dg)
(mod_cd.HSD <- TukeyHSD(mod_cd))
```

```r
adonis(all_dist_uc~Dg, data=meta_dist_uc, permutations=how(blocks=meta_dist_uc$Week, nperm=9999))
```

```r
mod_uc <- betadisper(all_dist_uc, meta_dist_uc$Dg)
(mod_uc.HSD <- TukeyHSD(mod_uc))
```
