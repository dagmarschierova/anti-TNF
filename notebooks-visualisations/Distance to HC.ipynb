```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dist_cd = pd.read_csv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_B/diversity_HCP_week038-CD/beta/bray_curtis_distance_matrix-modif.tsv", sep='\t')
dist_uc = pd.read_csv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_B/diversity_HCP_week038-UC/beta/bray_curtis_distance_matrix-modif.tsv", sep='\t')
#dist = pd.read_csv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_B/diversity_HCP_week038/beta/bray_curtis_distance_matrix-modif.tsv", sep='\t')
dist_cdf = pd.read_csv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_H/diversity_HCP_week038-CD/beta/bray_curtis_distance_matrix-modif.tsv", sep='\t')
dist_ucf = pd.read_csv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_H/diversity_HCP_week038-UC/beta/bray_curtis_distance_matrix-modif.tsv", sep='\t')
distf = pd.read_csv("/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_H/diversity_HCP_week038/beta/bray_curtis_distance_matrix-modif.tsv", sep='\t')
```

```python
dist_cd = pd.melt(dist_cd, value_vars=list(dist_cd.columns[0:]))
dist_uc = pd.melt(dist_uc, value_vars=list(dist_uc.columns[0:]))
#dist = pd.melt(dist, value_vars=list(dist.columns[0:]))
dist_cdf = pd.melt(dist_cdf, value_vars=list(dist_cdf.columns[0:]))
dist_ucf = pd.melt(dist_ucf, value_vars=list(dist_ucf.columns[0:]))
distf = pd.melt(distf, value_vars=list(distf.columns[0:]))
```

```python
f1, (ax1,ax2) = plt.subplots(1, 2, figsize=(20, 10))

sns.violinplot(ax=ax1, data=dist_cd, x='variable', y='value', color='white', inner='quartile')
sns.stripplot(ax=ax1, data=dist_cd, x="variable", y="value", size=10, alpha=.6, palette=['grey','orange'])
ax1.set_title("A", fontsize=30, loc='left')
ax1.set_ylabel('distance to HC', fontsize=25)
ax1.set_xlabel('')
ax1.tick_params(axis='both', labelsize=25)

sns.violinplot(ax=ax2, data=dist_uc, x='variable', y='value', color='white', inner='quartile')
sns.stripplot(ax=ax2, data=dist_uc, x="variable", y="value", size=10, alpha=.6, palette=['grey','red'])
ax2.set_title("B", fontsize=30, loc='left')
ax2.set_ylabel('distance to HC', fontsize=25)
ax2.set_xlabel('')
ax2.tick_params(axis='both', labelsize=25)

# statistical annotation
x1, x2 = 0, 1
y, h, col = dist_cd['value'].max() + .1, .01, 'k'
ax1.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
ax1.text((x1+x2)*.5, y+h, "p = 0.0001", ha='center', va='bottom', color=col, fontsize='xx-large')

x1, x2 = 0, 1
y, h, col = dist_uc['value'].max() + .1, .01, 'k'
ax2.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
ax2.text((x1+x2)*.5, y+h, "p = 0.0001", ha='center', va='bottom', color=col, fontsize='xx-large')

# save figure
f1.savefig('/Users/dagmarschierova/MEGA/MBÚ/TNF/NGS/runs_merged/outputs/qiime_results_B/diversity_HCP_week038/Distance_to_HC_CD_and_UC.png')
```
